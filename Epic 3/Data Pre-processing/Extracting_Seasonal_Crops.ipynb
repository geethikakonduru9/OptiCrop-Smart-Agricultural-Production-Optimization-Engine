```python
print("Summer Crops")
print(df[(df['temperature'] > 30) & (df['humidity'] > 50)]['label'].unique())

print("----------------------------------------")

print("Winter Crops")
print(df[(df['temperature'] < 20) & (df['humidity'] > 30)]['label'].unique())

print("----------------------------------------")

print("Rainy Crops")
print(df[(df['rainfall'] > 200) & (df['humidity'] > 50)]['label'].unique())
```
